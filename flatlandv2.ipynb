```python
# Import necessary libraries
from flatland.envs.flatland_env import FlatlandEnv
from flatland.envs.register import register_flatland_env
from flatland.core.env import Flatland
from flatland.envs.observations import TreeObsForRailEnv
from flatland.core.env import Env
import matplotlib.pyplot as plt

# Register a simple environment
register_flatland_env("FlatlandEnv-v0", FlatlandEnv)

# Create the environment with some parameters
env = FlatlandEnv(
    num_agents=2,           # Number of agents (trains)
    width=10,               # Width of the grid (number of columns)
    height=10,              # Height of the grid (number of rows)
    max_num_cities=2,       # Number of cities (stations)
    num_tracks=2,           # Number of tracks in the environment
    seed=42                 # Set a random seed for reproducibility
)

# Render the environment to visualize the layout
env.render(mode='human')  # 'human' mode is the default and renders the environment to screen

```
